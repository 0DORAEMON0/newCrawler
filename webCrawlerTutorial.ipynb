```python
import requests
from bs4 import BeautifulSoup
import csv
import time

keyDict = ['접수마감', '지원기간', '행사일정', '일시', '신청기간', '접수 기간', '접수기간', '모집기간']

links = []
boardNums = []
titles = []
components = []

def collectInfo(url):
    maxNum = checkMaxPages(url)
    setFeatureNames(features)
    spider(url, int(maxNum))

def checkMaxPages(url):
    webPage = requests.get(url)
    soup = BeautifulSoup(webPage.text, 'lxml')
    pageLists = soup.find('ul', 'pagination')
    temp = pageLists.find_all('li')
    maxNum = temp[-2].get_text().strip()
    
    return maxNum

def spider(url, max_pages):
    page = 1
    url = url + "?page="
    while page <= max_pages:
        try:
            web = requests.get(url + str(page))
        except:
            time.sleep(300)
            web = requests.get(url + str(page))
#         with requests.Session() as sess:
#             web = sess.get(url + str(page))
#             sess.cookies.clear()
        webPage = BeautifulSoup(web.text, 'lxml')
        getBoardNumber(webPage)
        getBoardTitles(webPage)
        getBoardLinks(webPage)
        for i in links:
            board = requests.get(i)
            content = BeautifulSoup(board.text, 'lxml')
            content = content.find('td', 'border_bottom board_content')
            extractWithKeyword(content)
#         print(links)
#         print(titles)
#         print(components)
#         print(boardNums)
        dataToFile(titles, components)
        links.clear()
        titles.clear()
        boardNums.clear()
        components.clear()
        print("page " + str(page) + " complete!")
        page += 1

def getBoardTitles(page):
    for i in page.select('tr > td > a'):
        titles.append(i.get_text().strip())

def getBoardLinks(page):
    for i in page.select('tr > td > a'):
        links.append(i['href'])

def getBoardNumber(page):
    cnt = 0
    for i in page.find_all('td', 'border_bottom txt_center b_num'):
        if cnt % 2 == 0:
            boardNums.append(i.get_text().strip())
        cnt += 1

def extractWithKeyword(contents):
    for i in keyDict:
        component = contents.find(text = lambda t: t and i in t)
        if component:
            break
#     print(component)
#     print(type(component))
    components.append(component)

features = ["게시판 번호", "제목", "기간정보"]
FILE_NAME = "boardInfo.csv"

def setFeatureNames(features):
    with open(FILE_NAME, 'w', encoding='utf-8', newline='') as f:
        f_writer=  csv.writer(f)
        f_writer.writerow(features)
    
def dataToFile(titles, components):
    with open(FILE_NAME, 'a', encoding='utf-8', newline='') as f:
        f_writer = csv.writer(f)
        length = len(titles)
        for i in range(length):
            f_writer.writerow([boardNums[i], titles[i], components[i]])
```
